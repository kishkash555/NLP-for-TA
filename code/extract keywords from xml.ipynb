```python

xmls = glob.glob('../articles_from_tatiana/batch3_header_xml/*.xml')

xmls[:10]
```

```python
for fl in xmls[5:10]:
    print(f'file: {fl}')
    print(list(get_keywords(fl)),end='\n\n')
```

```python
kws = [list(get_keywords(fl)) for fl in xmls]

flat_kws = sum(kws,[])

kw_counter = Counter(flat_kws)
```

```python
sorted(kw_counter.most_common(35))
```

```python
len(kw_counter)
```

```python
sorted([(k,v) for k,v in kw_counter.items() if v == 1])
```

```python
df =pd.DataFrame(data = list(kw_counter.items()), columns=['term','frequency']).sort_values('term')
```
