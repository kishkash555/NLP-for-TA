```python
nicknames = list(df.nickname)
```

```python
bibgraph = nx.Graph()

for name in nicknames:
    bibgraph.add_node(name)

for source_ind, jrf in enumerate(df.joint_reference_count):
    jrf = np.fromstring(jrf[1:-1],dtype=np.int, sep=',')
    for target_ind, weight in enumerate(jrf[source_ind+1:],source_ind+1):
        if weight > 0:
            bibgraph.add_edge(nicknames[source_ind], nicknames[target_ind], weight=weight)
```

```python
list(bibgraph.adjacency())
```

```python
write_gexf(bibgraph,'bibgraph_output.gexf')
```
