```python
from collections import Counter
```

```python
from os import path
import glob
```

```python
dr = '../articles_from_tatiana/batch3_citations/*'
fn = 'abishek2016.txt'

path.exists(path.join(dr,fn))
```

```python
fns = glob.glob(dr)
fns
```

```python
df = pd.DataFrame()
for fn in fns:
    curr = pd.read_csv(fn,sep='\t')
    curr.insert(0,'source',path.basename(fn)[:-4])
    df = df.append(curr)
df = df[['source','AU','PY','TI']]

df.shape
```

```python
ff = lambda s: [tuple(a.strip().split(',')) for a in s.split(';')] if type(s)==str else ('None','')
```

```python
df['author_list'] = df.AU.apply(func=ff)
```

```python
df['AU'][:10]
```

```python
df['author_list'][:10]
```

```python
df['first_auth_surname'] = df['AU'].map(lambda x: type(x)==str and x.split(';')[0].split(',')[0] or x)
df['year'] = df['PY']
df['title'] = df['TI']
```

```python
dl = []
for row_dict in df.to_dict(orient="records"):
    dl.append(bt.ref_dict_tools.add_nickname_field(row_dict))
```

```python
df1 = pd.DataFrame(dl)
```

```python
df1
```

```python
nick_count = Counter(df1.nickname)
df1['nick_count'] = df1.nickname.map(lambda nn: nick_count[nn])
print(df1.shape)
df2=df1[df1.nick_count>1]
print(df2.shape)
```

```python
df2.sort_values('nickname').to_excel('review_citations.xlsx',index=False)
```

```python
source_count = Counter(df2.source)
df2['source_count'] = df2.source.map(lambda sc: source_count[sc])
df3 = df2[df2.source_count>1]
df3.shape
```
