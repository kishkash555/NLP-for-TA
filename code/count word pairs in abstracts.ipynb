```python
def get_abstract_contents(fname):
    tree = etree.parse(fname)
    abstr = [a for a in tree.getiterator() if a.tag.endswith('abstract')]
    if abstr:
        text = ''.join([a.strip() for a in abstr[0].itertext()])
        return text
```

```python
def clean_words(s):
    ls = list(s)
    ls = [c for c in ls if c.isspace() or c.isalnum()]
    return ''.join(ls).lower()
```

```python
all_counters = {0: Counter(), 1: Counter(), 8: Counter(), -1: Counter()}

for file in xmls:
    key = path.basename(file)[4:-4]
    part = partition.get(key,-1) 
    try:        
        abstract_text = get_abstract_contents(file)
    except BaseException as e:
        print(f'error for {key}: {e}' )
        continue
    if len(abstract_text)<100:
        print("no abstract for {}".format(file))
        continue
    words = clean_words(abstract_text).split()
    pairs = list(zip(words,words[1:]))
    all_counters[part] += Counter(pairs)
```
