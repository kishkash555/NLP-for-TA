```python
for ref_list in b3_refs.values():
    for ref_entry in ref_list:
        if ref_entry.get('title','').startswith('Channel dynamics under') and ref_entry.get('year','')=='':
            ref_entry['year']='2000'
            bt.ref_dict_tools.add_nickname_field(ref_entry)
            print(ref_entry['title'], ref_entry.get('year'),ref_entry['nickname'])
```

```python
nickname_replacement_dict = {r.nickname: r.new_title for r in pdf[pdf.nickname!='__missing__'].itertuples()
                            if r.nickname != r.new_title}


for k, v in nickname_replacement_dict.copy().items():
    if v in ['tsay2004channel','tsay2009channel']:
        print(k,v, k in nickname_replacement_dict)
        del nickname_replacement_dict[k]
```

```python
c=0
for ref_list in b3_refs.values():
    for ref_entry in ref_list:
        if ref_entry['nickname'] in nickname_replacement_dict:
            old_nickname=ref_entry['nickname']
            new_nickname = nickname_replacement_dict[old_nickname]
            print('{} --> {}'.format(old_nickname, new_nickname))
            ref_entry['nickname'] = new_nickname
            c=c+1
print(c)
```
